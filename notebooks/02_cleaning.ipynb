```python
import sys
sys.path.append("../src")

from utils import load_all_data

dfs = load_all_data()
for key, df in dfs.items():
    print(f"\n{'='*50}\n {key}")
    
    # Doublons
    dupes = df.duplicated().sum()
    print(f"  Doublons     : {dupes}")
    
    # Types
    print(f"  Types :\n{df.dtypes.to_string()}")
    
    # Colonnes texte — vérifier la cohérence des catégories
    for col in df.select_dtypes(include=["object", "str"]).columns:
        uniq = df[col].unique()
        if len(uniq) < 20:  # Affiche seulement si peu de valeurs uniques
            print(f"  [{col}] → {list(uniq)}")
```
