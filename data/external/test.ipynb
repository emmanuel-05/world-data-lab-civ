```python
df_un = pd.read_csv("../external/un_population_civ.csv")

# Extraire les 3 indicateurs proprement
pop_15plus = df_un[
    df_un['IndicatorShortName'] == 'Population by age and sex - broad age groups'
][['Time', 'Value']].rename(columns={'Time': 'year', 'Value': 'pop_15plus'})

urb_rate = df_un[
    df_un['IndicatorShortName'] == 'Annual rate of population change by degree of urbanization'
][['Time', 'Value']].rename(columns={'Time': 'year', 'Value': 'urb_rate'})

urb_pct = df_un[
    df_un['IndicatorShortName'] == 'Percentage of annual rate of population change by degree of urbanization'
][['Time', 'Value']].rename(columns={'Time': 'year', 'Value': 'urb_pct'})

# Fusionner en un seul DataFrame propre
df_demo = pop_15plus \
    .merge(urb_rate, on='year') \
    .merge(urb_pct,  on='year')

df_demo['year'] = df_demo['year'].astype(int)
df_demo.to_csv("../external/demographics_clean_civ.csv", index=False)

print(df_demo)
```
